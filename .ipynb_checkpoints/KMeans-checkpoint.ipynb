```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
x1=np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2=np.array([5, 4, 6, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])
plt.plot()
plt.xlim([0,10])
plt.ylim([0,10])
plt.scatter(x1,x2)
plt.title("dataset")
plt.show()
X=np.array(list(zip(x1,x2))).reshape(len(x1),2)
colors=['r','b','g']
markers=['o','v','s']
K=3
k_clusters=KMeans(n_clusters=K).fit(X)
for i, label in enumerate(k_clusters.labels_):
    plt.plot(x1[i],x2[i],color=colors[label],marker=markers[label],ls='none')
plt.xlim([0,10])
plt.ylim([0,10])
plt.title("Clustered Dataset")
plt.show()
```
