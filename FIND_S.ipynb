```python
import pandas as pd
import numpy as np

data=pd.read_csv(".csv")
print(data)
d=np.array(data)[:,:-1]
print("the attributes are:",d)
target=np.array(data)[:,-1]
print("the target is:",target)

def train(c,t):
    s_h=None
    for i, val in enumerate(t):
        if val == "Yes":
            s_h=c[i].copy()
            break
    if s_h is None:
        print("there is no postive value")
        return None
    for i,val in enumerate(c):
        if t[i]=="Yes":
            for x in range(len(s_h)):
                if val[x]!=s_h[x]:
                    s_h[x]="?"
    return s_h
print("final output:",train(d,target))
```
